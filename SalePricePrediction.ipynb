```python
def wrangle(filename):
     # Read CSV file
    df = pd.read_csv(filename)

    #  we need to drop columns that are less that numbers/2
    cols=['Alley','FireplaceQu','PoolQC','Fence','MiscFeature']

    df.drop(columns=cols,axis=1,inplace=True)

    # After study of some columns let remove unnecessary and low cardinality columns
    cols1=['PavedDrive','Functional','Condition2','Condition1','GarageFinish','GarageQual','GarageCond','PavedDrive','Electrical','CentralAir','SaleType','BsmtFinType1','BsmtFinType2','Heating'
         ,'MSZoning','Street','Utilities','LotConfig','LandSlope','Id','KitchenAbvGr','GarageType','Fireplaces','BldgType']
    df.drop(columns=cols1,axis=1,inplace=True)

    # drop again some unnecessary columns
    cols2=['SaleCondition','YrSold','MoSold','ScreenPorch','3SsnPorch','EnclosedPorch','OpenPorchSF','WoodDeckSF','GarageYrBlt','TotRmsAbvGrd','KitchenQual','MSSubClass',
          'LotFrontage','OverallQual','OverallCond','YearRemodAdd','BsmtFinSF2','MasVnrArea','WoodDeckSF','PoolArea','MiscVal','LotShape','ExterQual','ExterCond']
    df.drop(columns=cols2,axis=1,inplace=True)

    #  drop again some unnecessary columns
    cols4=['RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinSF1','BsmtUnfSF','HeatingQC','1stFlrSF','2ndFlrSF','LowQualFinSF','GarageCars']
    df.drop(columns=cols4,axis=1,inplace=True)
#      Drop columns with high miss value 
    df.drop(columns=['floor','expenses'],axis=1,inplace=True)

    #  drop columns that the corelation is > 50
    cols5=['BsmtHalfBath','FullBath','GarageArea','BedroomAbvGr','BsmtFullBath','HalfBath']
    df.drop(columns=cols5,axis=1,inplace=True)
    
    
    return df

```

```python
from category_encoders import OneHotEncoder
from sklearn.linear_model import LinearRegression, Ridge 
from sklearn.metrics import mean_absolute_error
from sklearn.pipeline import make_pipeline
```

```python
df.columns
```

```python
# split the data into target and feature

X_train=df.drop('SalePrice',axis=1)
target = "SalePrice"

y_train=df[target]
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

## Build The Model

```python
# Calculate the baseline
y_mean=y_train.mean()
y_pred_baseline=[y_mean]*len(y_train)
print("Mean apt price:",y_mean )

print("Baseline MAE:", mean_absolute_error(y_train,y_pred_baseline))
```

```python
# Iterate
model = make_pipeline(OneHotEncoder(use_cat_names=True),
                    
                      Ridge()
                     )
model.fit(X_train,y_train)
```

## Evaluate


```python
y_pred_traning=model.predict(X_train)
print("Training MAE:", mean_absolute_error(y_train,y_pred_traning))
```
